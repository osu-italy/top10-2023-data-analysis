```python
import json
out_data = {}
```

```python
with open("anonymized_and_cleaned_votes.json") as f:
    data = json.load(f)
```

```python
for player_vote in data:
    tierlist = player_vote["votes"]
    for vote in tierlist:
        voted = vote["choiceUsername"]
        if voted not in out_data:
            out_data[voted] = 0
        out_data[voted] += vote["weight"]
```

```python
import pandas

df = (
    pandas.DataFrame(out_data.items(), columns=["username", "points"])
    .sort_values("points", ascending=False)
    .reset_index(drop=True)
)

# only the first 10 values
top10 = df.head(10)
top10.index += 1
top10
```
